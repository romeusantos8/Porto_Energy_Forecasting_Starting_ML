```python
import pandas as pd

df = pd.read_csv("../data/raw/weather_data_2019-01-01_to_2023-12-31.csv", parse_dates=["time"])
df.head()

df["hour"] = df["time"].dt.hour
df["dayofweek"] = df["time"].dt.dayofweek
df["month"] = df["time"].dt.month

df_daylight = df[df["shortwave_radiation"] > 0].copy()

df_daylight["radiation_lag_1"] = df_daylight["shortwave_radiation"].shift(1)
df_daylight["radiation_lag_24"] = df_daylight["shortwave_radiation"].shift(24)

y = df_daylight["shortwave_radiation"]

features = [
    "temperature_2m",
    "cloudcover",
    "hour",
    "month",
    "radiation_lag_1",
    "radiation_lag_24"
]

X = df_daylight[features]

X = X.dropna()
y = y.loc[X.index]

X.shape, y.shape

X.head(), y.head()

```
